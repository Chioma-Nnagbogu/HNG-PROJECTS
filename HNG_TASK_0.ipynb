```python
colors = sns.color_palette('Set2', n_colors=len(df['COUNTRY'].unique()))
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x='COUNTRY', y='SALES', data=df, ci=None, palette=colors)
plt.title('Sales by Country')
plt.xlabel('Country')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='QTR_ID', y='QUANTITYORDERED', data=df,palette=colors )
plt.title('Order Quantity per Quarter' )
plt.xlabel('Quarter')
plt.ylabel('Order Quantity')
plt.grid(True)
plt.show()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df['Year'] = df['DATE'].dt.year
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='SALES',  data=df, marker='o', markersize=10,palette=colors)
plt.title('Sales vs Year ')
plt.xlabel('Year')
plt.ylabel('SALES')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x='COUNTRY', y='QUANTITYORDERED', data=df, ci=None,palette=colors)
plt.title('Order Quantity per COUNTRY' )
plt.xlabel('COUNTRY')
plt.ylabel('QUANTITYORDERED')
plt.grid(True)
plt.show()
```
